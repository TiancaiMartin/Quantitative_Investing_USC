## Problem 3

```python
# Load data
spy_df = pd.read_csv('SPY_HW1.csv', skiprows = 2, index_col='Date', parse_dates=True)
spy_df.columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
```

### Problem 3.a

```python
import numpy as np
# Compute the daily returns over the daytime period (open-to-close) of SPY and add them as a new column to the dataframe
spy_df['Daily Return (open-to-close)'] = np.log(spy_df['Close'] / spy_df['Open'])
# Compute the daily returns over the nighttime period (close-to-open) of SPY and add them as a new column to the dataframe
spy_df['Daily Return (close-to-open)'] = np.log(spy_df['Open'] / spy_df['Close'].shift(1))
# Display the average of the two daily return series
average_open_to_close = spy_df['Daily Return (open-to-close)'].mean()
average_close_to_open = spy_df['Daily Return (close-to-open)'].mean()
print(f"Average Daily Return (open-to-close): {average_open_to_close * 100:.4f}%")
print(f"Average Daily Return (close-to-open): {average_close_to_open * 100:.4f}%")
```

### Problem 3.b

```python
# Compute the average full-day return using close-to-close prices
spy_df['Full-Day Return (close-to-close)'] = np.log(spy_df['Close'] / spy_df['Close'].shift(1))
average_full_day_return = spy_df['Full-Day Return (close-to-close)'].mean()
# Display the average full-day return
print(f"Average Full-Day Return (close-to-close): {average_full_day_return * 100:.4f}%")
```

### Problem 3.c

```python
# Compute the average full-day return using adjusted close-to-adjusted close prices
spy_df['Full-Day Return (adj close-to-adj close)'] = np.log(spy_df['Adj Close'] / spy_df['Adj Close'].shift(1))
average_full_day_return_adj = spy_df['Full-Day Return (adj close-to-adj close)'].mean()
# Display the average full-day return using adjusted close prices
print(f"Average Full-Day Return (adj close-to-adj close): {average_full_day_return_adj * 100:.4f} %")
```

## <mark> Problem solution summary: <mark>

From the above questions in **problem 3**, we load the SPY data from the CSV file into a pandas dataframe. We then compute the daily log returns for both the daytime (open-to-close) and nighttime (close-to-open) periods by using numpy.log. We also apply the same method to compute the full-day log returns using both close-to-close prices and adjusted close-to-adjusted close prices. Finally, we calculate and display the average returns for each of these periods.

### Problem 3.d

The difference between close-to-close returns (using the raw closing price) and adjusted close-to-adjusted close returns arises because _**the adjusted close accounts for dividends, stock splits, and other adjustments**_. The regular close price only reflects the last traded price of the day, while the adjusted close price back-adjusts the historical series to ensure returns are measured on total-return basis.
